```python
#Write a Python program to find words which are greater than given length k?

def wordlen(a,k):
    z=a.split()
    for i in range(0,len(z)):
        if int(len(z[i]))>int(k):
              print(z[i])
    return None
              
word=input("enter your words")
k=int(input("enter the max length og word"))

wordlen(word,k)
```

```python
#Write a Python program for removing i-th character from a string?

def char(a,k):
    new_str=" "
    for i in range(len(a)):
        if i==k:
            continue
        else:
            new_str=new_str+a[i]
    print(new_str)
               
word=input("enter your words")
k=int(input("enter the max length of word"))

char(word,k)
```

```python
#Write a Python program to split and join a string?

def spljoin(a):
    z=list(a)
    print("after splitting the str will be {}".format(z))
    new_str=" "
    for i in z:
        new_str=new_str+i
    print("after again joining the str we get {}".format(new_str))
    
    
spljoin("hsgjagd")  
```

```python
#Write a Python to check if a given string is binary string or not?

def bin(a):
    count=0
    for i in a:
        if int(i) not in range(0,2):
            count=1
        else:continue
    if count==1:
        print("not an binary string")
    else:
        print("binary string")

number=input("enter string")
bin(number)
```

```python
#Write a Python program to find uncommon words from two Strings?

def uncommon(a,b):
    d=[]
    for i in a:
        n=len(b)
        count=0
        while(n>=0):
            if str(i)==str(b[n-1]):
                count=1
                break
            else:
                count=0
                n=n-1
                continue
        if count!=1:
            d.append(i)
        else:continue
    print(d)

a=input("enter your 1 sttring")
b=input("enter your 2 string")
uncommon(a,b)
```
