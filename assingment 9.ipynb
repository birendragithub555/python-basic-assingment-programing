```python
#Write a Python program to print all disarium numbers between 1 to 100?
def diss(a):
    c=str(a)
    d=list(c)
    w={}
    sum=0
    for ind,i in enumerate(d):
        w[ind+1]=int(i)
    for k,v in w.items():
        sum=sum+v**k
    if int(sum)==int(a):
        print("{} is a disarium number".format(sum))

for i in range(1,101):
    diss(i)
```

```python
#Write a Python program to check if the given number is Happy Number?
def happy(num):
    sum=0
    while(num>0):
        a=num%10
        sum=sum+a*a
        num=num//10
    return sum

num=int(input('enter your number'))
result=happy(num)

while(result!=1 and result!=4):
    result=happy(result)
if result==1:
    print("happy number")
elif result==4:
    print("unhappy number")


```

```python
#Write a Python program to print all happy numbers between 1 and 100?

def happy(num):
    sum=0
    while(num>0):
        a=num%10
        sum=sum+a*a
        num=num//10
    return sum

for i in range(1,101):
    result=i
    while(result!=1 and result!=4):
        result=happy(result)
    if result==1:
        print("{} is a happy number".format(i))
```

```python
#Write a Python program to determine whether the given number is a Harshad Number?

def harshad(num):
    sum=0
    while(num>0):
        a=num%10
        sum=sum+a
        num=num//10
    return sum

number=int(input("enter the number"))
c=number
d=harshad(number)

if  c%d==0:
    print("harshad number")
else:
    print("not an harshad number")
```
