```python
#Write a Python program to find the sum of all items in a dictionary?

dict = {'511':'Vishnu','512':'Vishnu','513':'123','514':'Ram','515':'sita'}
sum1=0
for k,v in dict.items():
    for i in k:
        if int(i) in range(0,9):
            sum1=sum1+int(k)
            break
    for j in v:
        try:
            if int(j) in range(0,9):
                sum1=sum1+int(v)
                break
        except:
            ValueError
            continue
print(sum1)   
```

```python
#Write a Python program to Merging two Dictionaries?
dict1 = {'511':'Vishnu','512':'Vishnu','513':'123','514':'Ram','515':'sita'}
dict2 = {'516':'abd','517':'cnd','518':'126','519':'m','520':'gstag'}

dict3={}
for k,v in dict1.items():
    dict3[k]=v
for s,g in dict2.items():
    dict3[s]=g
dict3
```

```python

#Write a Python program to insertion at the beginning in OrderedDict?

from collections import OrderedDict as od
s=od([("a",1),("b",2)])
s.update({"d":4})
s.move_to_end("d",last=False)
s
```
