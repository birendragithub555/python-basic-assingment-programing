```python
#Write a Python Program to Print the Fibonacci sequence?
c=int(input("enter your number of counts for fibonacci series"))
n1=0
n2=1
if c<=0:
    print("enter positive number")
else:
    count=0
    if c==1:
        print(n1)
    elif c==2:
        print(n1)
        print(n2)
    else:
        print(n1)
        print(n2)
        while(count<c-2):
            n=n1+n2
            print(n)
            n1=n2
            n2=n
            count+=1
```

```python
#Write a Python Program to Check Armstrong Number?
number4=input("enter your number")
c=[]
b=0
for i in number4:
    c.append(i)
for k in range(len(c)):
    a=int(c[k])*int(c[k])*int(c[k])
    b=b+a
print(b)
if int(b)==int(number4):
    print("after cubing each letter we get {} as Armstrong Number".format(number4))
else:print("after cubing each letter we get {} as not an Armstrong Number".format(number4))
```

```python
#Write a Python Program to Find Armstrong Number in an Interval?
range_1=int(input("enter your number"))
range_2=int(input("enter your number"))
e=[]
for i in range(int(range_1),int(range_2+1)):
    e.append(i)
for j in e:
    m=j
    sum=0
    while(j>0):
        c=j%10
        sum=sum+c*c*c
        j=j//10
    if int(sum)==int(m):
        print("{} is armstrong number".format(m))
    else:
        continue  
```
