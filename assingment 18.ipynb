```python
#Create a function that takes a list of non-negative integers and strings and return a new listwithout the strings.

def filter(a):
    l=[]
    for i in a:
        if type(i)==int:
            l.append(i)
        else:continue
    return l
filter([1,"abc","123","dubai",2,14])
```

```python
#The "Reverser"; takes a string as input and returns that string in reverse order, with theopposite case.

def reverse(a):
    l=[]
    b=a.swapcase()
    for i in range(len(b)):
        l.append(b[len(b)-i-1])
    c="".join(l)
    return c

reverse("asddSWQW")
```

Write a function that calculates the factorial of a number recursively.

```python
def factorial(a):
    if a==0:
        return 1
    else:
        return a*factorial(a-1)
factorial(5)
```

Write a function that moves all elements of one type to the end of the list.

```python
def move_end(a,b):
    l=[]
    d=[]
    for i in a:
        if i==b:
            l.append(i)
        else:d.append(i)
    for i in l:
        d.append(i)
    return d

move_end([1,2,3,4,5,5,6],2)
    
```
