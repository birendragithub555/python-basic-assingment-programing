```python
Write a Python Program to find sum of array?

def sum1(*args):
    sum=0
    for i in args:
        for j in i:
            sum=sum+j
    return sum
sum1([1,2,3])
```

```python
#write a Python Program to find largest element in an array?

def large(*args):
    temp=0
    for i in args:
        for j in i:
            if int(j)>int(temp):
                temp=j
        else:continue
    return temp

large([78,37864872347,0])
```

```python
#Write a Python Program for array rotation?
def rotation(arr, n, d):
    temp = []
    i = 0
    while (i < d):
        temp.append(arr[i])
        i = i + 1
    i = 0
    while (d < n):
        arr[i] = arr[d]
        i = i + 1
        d = d + 1
    arr[:] = arr[:i] + temp
    return arr
 
arr = [1, 2, 3, 4, 5, 6, 7]
print("Array after left rotation is: ", end=' ')
print(rotation(arr, len(arr), 2))
```

```python
#Write a Python Program to Split the array and add the first part to the end?

def split(arr2,n1,d1):
    b=arr2[:d1]
    arr2[:]=arr2[d1:]+b
    return arr2
arr2=[1,2,3,4,5]
n1=len(arr2)
d1=4
split(arr2,n1,d1)
```

```python
#Write a Python Program to check if given array is Monotonic?

def mono(a):
    b=[]
    for i in a:
        b.append(i)
    if (sorted(b)==b):
            print("monotonus")
    elif (sorted(b,reverse=True)==b):
            print("monotonus")
    else:print("non monotonous")
    return None 
mono([1,4,5,2])
```
