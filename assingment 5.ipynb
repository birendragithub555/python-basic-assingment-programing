```python
#Write a Python Program to Find LCM?
number1=int(input("enter yor number"))
number2=int(input("enter yor number"))
biggest=0
if number1>number2:
    biggest=number1
else:biggest=number2
while(True):
    if (biggest%number1)==0 and (biggest%number2)==0:
        print("lcm is {}".format(biggest))
        break
    else:
        biggest=biggest+1
        
```

```python
#Write a Python Program to Find HCF?
number1=int(input("enter your number"))
number2=int(input("enter your number"))
smallest=0
if number1>number2:
    smallest=number2
else:smallest=number1
while(True):
    if (number1%smallest)==0 and (number2%smallest)==0:
        print("hcf={}".format(smallest))
        break
    else:smallest=smallest-1
```

```python
#Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?
decimal=int(input("enter your decimal number"))
print("binary number of {} is:{}".format(decimal,bin(decimal)))
print("octal number of {} is:{}".format(decimal,oct(decimal)))
print("hexadecimal number of {} is:{}".format(decimal,hex(decimal)))
```

```python
#Write a Python Program To Find ASCII value of a character?
asci=input("enter your character")
print("ascii value of {} is:{}".format(asci,ord(asci)))
```
