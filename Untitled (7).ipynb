```python
#Write a Python Program to find sum of array?

def sum1(*args):
    sum=0
    for i in args:
        for j in i:
            sum=sum+j
    return sum

sum1([1,2,3])
```

```python
#Write a Python Program to find largest element in an array?

def large(*args):
    temp=0
    for i in args:
        for j in i:
            if int(j)>int(temp):
                temp=j
        else:continue
    return temp

large([78,37864872347,0])
```

```python
#Write a Python Program for array rotation?

def rotation(arr, n, d):
    temp = []
    i = 0
    while (i < d):
        temp.append(arr[i])
        i = i + 1
    i = 0
    while (d < n):
        arr[i] = arr[d]
        i = i + 1
        d = d + 1
    arr[:] = arr[:i] + temp
    return arr
 
arr = [1, 2, 3, 4, 5, 6, 7]
print("Array after left rotation is: ", end=' ')
print(rotation(arr, len(arr), 2)) 
```
