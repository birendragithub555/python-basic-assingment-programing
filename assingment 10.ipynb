```python
#Write a Python program to find sum of elements in list?

no=int(input("enter the number of elements in array"))
d=[]
for i in range(0,no):
    ele=int(input("enter your element"))
    d.append(ele)
print("The sum of your array {} is {}".format(d,sum(d)))
```

```python
#Write a Python program to Multiply all numbers in the list?

no1=int(input("enter the number of elements in array"))
e=[]
for i in range(0,no1):
    ele1=int(input("enter your element"))
    e.append(ele1)
mul=1
for j in e:
    mul=mul*j
print("the multiplication of array {} is {}".format(e,mul))
```

```python
# Write a Python program to find smallest number in a list?

small=int(input("enter the number of elements in array"))
f=[]
for i in range(0,small):
    ele2=int(input("enter your element"))
    f.append(ele2)
temp=f[0]
for i in f:
    if i<temp:
        temp=i
    else:continue
print(temp)
```

```python
#Write a Python program to find second largest number in a list?
large=int(input("enter the number of elements in array"))
f=[]
for i in range(0,large):
    ele2=int(input("enter your element"))
    f.append(ele2)
f.sort(reverse=True)
print("second largest number in a list {} is {}".format(f,f[1]))
```

```python
#Write a Python program to find N largest elements from a list?
s=int(input("enter the number of elements in array"))
h=[]
for i in range(0,s):
    ele2=int(input("enter your element"))
    h.append(ele2)
n=int(input("enter your N number"))
print(h)

d=[]
def nlarge(li,n):
        li.sort(reverse=True)
        for i in range(0,n):
            d.append(li[i])
        return d
            

nlarge(h,n)
```

```python
#Write a Python program to print even numbers in a list?

s=int(input("enter the number of elements in array"))
h=[]
for i in range(0,s):
    ele2=int(input("enter your element"))
    h.append(ele2)
print(h)
d=[]
for i in h:
    if i%2==0:
        d.append(i)
    else:continue
print("the even numbers in list are {}".format(d))


```

```python
#Write a Python program to Remove empty List from List?

def empty(*args):
    s=[]
    for i in list(args):
        if i==[]:
            continue
        else:
            s.append(i)
    print(s)

empty([],[1,2,3],[])

```

```python
#Write a Python program to Cloning or Copying a list?

def Cloning(li1):
    li_copy = li1[:]
    return li_copy

li1 = [4, 8, 2, 10, 15, 18]
li2 = Cloning(li1)
print("Original List:", li1)
print("After Cloning:", li2)
```
