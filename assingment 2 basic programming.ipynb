```python
#Q1 Write a Python program to convert kilometers to miles?

A=int(input("enter your km"))
print("your distance in miles is {}".format(A*0.621371))
```

```python
#Q2 Write a Python program to convert Celsius to Fahrenheit?

A=int(input("enter your temperature in degree celsius"))
print("your temp in fahrenheit is {}".format((A*(9/5)) + 32))
```

```python
#Q3 Write a Python program to display calendar?
import calendar
month=int(input("enter your month"))
year=int(input("enter your year"))
print(calendar.month(year,month))
```

```python
#Q4 Write a Python program to solve quadratic equation?
import cmath
a=int(input("enter number a"))
b=int(input("enter number b"))
c=int(input("enter number c"))
d=b*2-4*a*c
if b*2-4*a*c==0:
    print("infinity condition")
else:
    x1=(-b-cmath.sqrt(d)/(2*a))
    x2=(-b+cmath.sqrt(d)/(2*a))
    print("values of x1 ={} and x2={}".format(x1,x2))
    
```

```python
#Q5 Write a Python program to swap two variables without temp variable?
a=int(input("enter number a"))
b=int(input("enter number b"))
c=[a,b]
d=[]
for i in range(len(c)):
    d.append(c[len(c)-i-1])
print("the values after swapping  of a={} and b={}".format(d[0],d[1]))
    

```
